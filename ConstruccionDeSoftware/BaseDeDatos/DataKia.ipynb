# [dbo].[iniciarSesion]

```sql
/****** Object:  StoredProcedure [dbo].[iniciarSesion]    Script Date: 29/08/2024 08:52:06 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[iniciarSesion]
    @Usuario NVARCHAR(100),
    @Contrasena NVARCHAR(100)
AS
BEGIN
    -- Lógica para iniciar sesión del usuario
    DECLARE @Exito BIT;
    SET @Exito = (SELECT CASE WHEN EXISTS (
        SELECT 1 FROM Usuarios WHERE Nombre = @Usuario AND Contrasena = @Contrasena
    ) THEN 1 ELSE 0 END);
    
    IF @Exito = 1
    BEGIN
        PRINT 'Inicio de sesión exitoso';
    END
    ELSE
    BEGIN
        PRINT 'Credenciales incorrectas';
    END
END;
GO
```
